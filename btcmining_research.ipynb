```python
#import libraries
import requests
import pandas as pd
import datetime as dt
import coinmetrics
import plotly.graph_objects as go
import numpy as np
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
# Initialize a reference object for the Community API
cm = coinmetrics.Community()

# List the assets Coin Metrics has data for
supported_assets = cm.get_supported_assets()
print("supported assets:\n", supported_assets)

# List all available metrics for BTC
asset = "btc"
available_data_types = cm.get_available_data_types_for_asset(asset)
print("available data types:\n", available_data_types)

# Obtain `HashRate` and `BlkCnt` data for BTC from 2017-01-01 to 2021-08-18.
asset = "btc"
metric = "HashRate,BlkCnt"
begin_timestamp = "2017-01-01"  # The `datetime` type is also accepted
end_timestamp = "2021-08-18"  # The `datetime` type is also accepted
asset_data = cm.get_asset_data_for_time_range(asset, metric, begin_timestamp, end_timestamp)
print("data given timerange and metric:\n", asset_data)
```

```python
# Convert the data object to a Pandas DataFrame for further processing
mining_data_df = coinmetrics.cm_to_pandas(asset_data)
```

```python
# remove time from DatetimeIndex
mining_data_df.index = pd.DatetimeIndex(mining_data_df.index)
mining_data_df.index = mining_data_df.index.date
```

```python
# Save the Pandas DataFrame OR raw Coimetrics object into a CSV
path_or_filename = "gdrive/My Drive/output.csv"
coinmetrics.csv(mining_data_df, path_or_filename)
print("CSV written:", path_or_filename)
```

```python
# plot a line graph using plotly
Date = mining_data_df.index
Hash_Rate = mining_data_df['HashRate']
fig = go.Figure(
    data=[go.Scatter(x= Date, y=Hash_Rate)],
    layout_title_text="Hash Rate (TH/s):")

fig.show()
fig.write_html('gdrive/My Drive/btc_mining_research_hr')
```

```python
# plot a line graph using plotly
Date = mining_data_df.index
Block_Count = mining_data_df['BlkCnt']
fig = go.Figure(
    data=[go.Scatter(x= Date, y=Block_Count)],
    layout_title_text="Block Count:"
)
fig.show()
fig.write_html('gdrive/My Drive/btc_mining_research_BlkCnt')
```
